```python
# 2) (Warning: This exercise is for people who are familiar with statistics.) Loop over all the variables in the dataset,
#    and perform an independent-samples t-test on the data for the binarized wine quality. Which variables are significantly
#    different between "low" and "high" quality wine?

# separate into two groups
low_quality = data[data['boolQuality'] == 0]
high_quality = data[data['boolQuality'] == 1]

# Store results
results = []

for col in data.columns:
  if col not in ['quality', 'boolQuality']:
    t_stat, p_val = stats.ttest_ind(low_quality[col], high_quality[col], equal_var=False)
    results.append((col, t_stat, p_val))

# Filter significant variables
significant_vars = [col for col, t, p in results if p < 0.05]
```

```python
# Create a class for the model
class ANNWine(nn.Module):

  def __init__(self):

    super().__init__()

    ### input layer
    self.input = nn.Linear(11, 16)

    ## hidden layers
    self.fc1 = nn.Linear(16, 32)
    self.fc2 = nn.Linear(32, 32)

    ## output layer
    self.output = nn.Linear(32, 1)

  def forward(self, x):
    x = F.relu( self.input(x) )
    x = F.relu( self.fc1(x) )
    x = F.relu( self.fc2(x) )
    return self.output(x)
```

```python
import numpy as np
import torch

class EarlyStopping:
    def __init__(self, patience=5, delta=0, path='checkpoint.pt', verbose=False):
        self.patience = patience
        self.delta = delta
        self.path = path
        self.verbose = verbose

        self.counter = 0
        self.best_score = None
        self.early_stop = False
        self.val_loss_min = np.inf

    def __call__(self, val_loss, model):

        score = -val_loss

        if self.best_score is None:
            self.best_score = score
            self.save_checkpoint(val_loss, model)
        elif score < self.best_score + self.delta:
            self.counter += 1
            if self.verbose:
                print(f'EarlyStopping counter: {self.counter} out of {self.patience}')
            if self.counter >= self.patience:
                self.early_stop = True
        else:
            self.best_score = score
            self.save_checkpoint(val_loss, model)
            self.counter = 0

    def save_checkpoint(self, val_loss, model):
        '''Saves model when validation loss decreases.'''
        if self.verbose:
            print(f'Validation loss decreased ({self.val_loss_min:.6f} --> {val_loss:.6f}). Saving model...')
        torch.save(model.state_dict(), self.path)
        self.val_loss_min = val_loss
```

```python
# A function that trains the model

# global parameter
num_epochs = 1000

def trainTheModel(model, num_epochs=num_epochs):

  # loss function and optimizer
  lossfun = nn.BCEWithLogitsLoss()
  optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

  # initialise losses
  losses = torch.zeros(num_epochs)
  trainAcc = []
  testAcc  = []

  # Loop over epochs
  for epochi in range(num_epochs):

    # switch on training mode
    model.train()

    # loop over training data batches
    batchAcc = []
    batchLoss = []

    for X, y in train_loader:

      # forward pass and loss
      yhat = model(X)
      loss = lossfun(yhat, y)

      # backprop
      optimizer.zero_grad()
      loss.backward()
      optimizer.step()

      # loss from this batch
      batchLoss.append(loss.item())

      # compute training accuracy for this batch
      batch_acc = 100 * torch.mean(((yhat > 0) == y).float()).item()

      batchAcc.append(batch_acc)
    # end of batch loop

    # now that we've trained through the batches, get their average training accuracy
    trainAcc.append( np.mean(batchAcc) )

    # and get average losses acrpss the batches
    epoch_loss = np.mean(batchLoss)
    losses[epochi] = epoch_loss

    # early_stopping(epoch_loss, model)

    # if early_stopping.early_stop:
    #     print("Early stopping")
    #     break

    # test accuracy
    model.eval()
    X, y = next(iter(test_loader)) # extract X, y from the test dataloader
    with torch.no_grad():
      yhat = model(X)
    testAcc.append( 100*torch.mean(((yhat > 0) == y).float()).item() )


  # function output
  return trainAcc, testAcc, losses
```

```python
# range of batch sizes
batchsizes = 2**np.arange(1, 10, 2)

# initialise output results matrices
accuracyResultsTrain = np.zeros((num_epochs, len(batchsizes)))
accuracyResultsTest  = np.zeros((num_epochs, len(batchsizes)))
comptime             = np.zeros(len(batchsizes))

# loop over batch sizes
for bi in range(len(batchsizes)):

  # start the clock!
  starttime = time.process_time()

  # create dataloader object
  train_loader = DataLoader(train_data, batch_size=int(batchsizes[bi]), shuffle=True, drop_last=True)

  # create and train a model
  winenet = ANNWine()
  trainAcc, testAcc, losses = trainTheModel(winenet)

  # store only up to the length of the returned accuracies
  epochs_ran = len(trainAcc)
  accuracyResultsTrain[:epochs_ran, bi] = trainAcc
  accuracyResultsTest[:epochs_ran, bi] = testAcc

  # check the timer
  comptime[bi] = time.process_time() - starttime

# import numpy as np
# import time
# from torch.utils.data import DataLoader

# # Define batch sizes to test
# batchsizes = 2**np.arange(1, 10, 2)  # [2, 8, 32, 128, 512]

# # Set number of epochs for pre-allocation (max possible)
# num_epochs = 1000

# # Preallocate result arrays
# accuracyResultsTrain = np.zeros((num_epochs, len(batchsizes)))
# accuracyResultsTest  = np.zeros((num_epochs, len(batchsizes)))
# comptime             = np.zeros(len(batchsizes))

# # Loop over different batch sizes
# for bi, batch_size in enumerate(batchsizes):

#     print(f"\n--- Training with batch size: {batch_size} ---")

#     # Start the timer (wall time, better for GPU)
#     starttime = time.time()

#     # Create DataLoader with the current batch size
#     train_loader = DataLoader(train_data, batch_size=int(batch_size), shuffle=True, drop_last=True)

#     # Initialize model
#     winenet = ANNWine()

#     # Train the model (you must ensure this function uses train_loader and test_loader correctly)
#     trainAcc, testAcc, losses = trainTheModel(winenet)

#     # Store results only if output lengths match and are valid
#     epochs_ran = len(trainAcc)

#     if epochs_ran > 0 and len(testAcc) == epochs_ran:
#         accuracyResultsTrain[:epochs_ran, bi] = trainAcc
#         accuracyResultsTest[:epochs_ran, bi] = testAcc
#     else:
#         print(f"⚠️ Warning: Skipped storing results for batch size {batch_size} due to mismatch or empty output.")
#         continue

#     # Store computation time
#     comptime[bi] = time.time() - starttime

#     print(f"⏱️ Training time: {comptime[bi]:.2f} seconds")

```
