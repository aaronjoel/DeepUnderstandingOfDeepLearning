```python
r = 0.025   # interest rate
T = 50      # end date
b = np.empty(T+1)  # an empty NumPy array, to store all b_t
b[0] = 10          # initial balance

for t in range(T):
  b[t+1] = (1 + r) * b[t]    # b3 = (1 + r)*b2 = (1 + r)*(1 + r)*b1 = (1 + r)*(1 + r)*(1 + r)*b0 = (1 + r)^3 * b[0]

plt.plot(b, label='Bank balance')
plt.legend()
plt.show()
```

```python
# Simulate and plot the correlated time series
# x_t+1 = alpha*x_t + e_t+1, where x0 = 0, t = 0, ...T

import numpy as np
import matplotlib.pyplot as plt

T = 200
alpha = 0.9
x = np.empty(T+1)
x[0] = 0

e_values = np.random.randn(T+1)

for t in range(T):
  x[t+1] = alpha * x[t] + e_values[t+1]

plt.plot(x, label='Correlated time series')
plt.legend()
plt.show()
```

```python
n = 1000000  # sample size for Monte Carlo simulation

count = 0
for i in range(n):

  # drawing random positions on the square
  u, v = np.random.uniform(), np.random.uniform()

  # Check whether the point falls within the boundary
  # of the unit circle centered at (0.5, 0.5)
  d = np.sqrt((u - 0.5)**2 + (v - 0.5)**2)

  # if it falls within the inscribed circle,
  # add it to the count
  if d < 0.5:
    count += 1

area_estimate = count / n
```

```python
def generate_data(n, generator_type):
  e_values = []
  for i in range(n):
    if generator_type == 'U':
      e = np.random.uniform(0, 1)
    else:
      e = np.random.randn()
    e_values.append(e)

  return e_values

data = generate_data(100, 'U')
plt.plot(data)
plt.show()
```

```python
def factorial_loop(n):
  result = 1
  for i in range(1, n+1):
    result *= i
  return result
```

```python
def factorial(n):
  if n == 0:
    return 1
  elif n == 1:
    return 1
  else:
    return n * factorial(n-1)
```

```python
from numpy.random import uniform

def binomial_rv(n, p):
  count = 0
  for i in range(n):
    U = uniform()
    if U < p:
      count += 1
  return count

```

```python
binomial_rv(10, 0.5)
```

```python
def draw(k):
  payoff = 0
  count = 0

  for i in range(10):
    U = uniform()
    count = count + 1 if U < 0.5 else 0
    print(f'count: {count}')
    if count == k:
      payoff = 1

  return payoff
```

```python
draw(3)
```

```python
draw(4)
```

```python
def draw_new(k):  # pays if k successes in a sequence

    payoff = 0
    count = 0

    for i in range(10):
        U = uniform()
        count = count + ( 1 if U < 0.5 else 0 )
        print(count)
        if count == k:
            payoff = 1

    return payoff

draw_new(3)
```
